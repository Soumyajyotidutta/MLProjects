```python
# input texts
a = "flying fish flew by the space station"
b = "Netflix produced this super cool show called 'Squid Game' and I love it"
c = "Let's go skating to colorado, I am sure that will be super fun"
d = "Netflix prodcued this super cool show called 'Sweet Home' and I love it"
```

```python
# shingle function
def shingle(text: str, k: int):
  shingle_set = []
  for i in range(len(text) - k + 1):
    shingle_set.append(text[i:i+k])
  return sorted(list(shingle_set))
```

```python
k = 2
a = shingle(a, k)
b = shingle(b, k)
c = shingle(c, k)
d = shingle(d, k)
print(a)
```

```python
vocab = a + b + c + d
print(vocab)
```

```python
# one hot coding
a1Hot = [1 if x in a else 0 for x in vocab]
b1Hot = [1 if x in b else 0 for x in vocab]
c1Hot = [1 if x in c else 0 for x in vocab]
d1Hot = [1 if x in d else 0 for x in vocab]
print(a1Hot)
```

```python
from random import shuffle
```

```python
# minhashing (convert sparse vectors into dense vactors/signatures)

# 1. function for creating hash vector
def createHashFunction(size: int):
  hashEx = list(range(1, len(vocab)+1))
  shuffle(hashEx)
  return (hashEx)

# 2. function for creating multiple minhash vectors
def createMinhashFunction(vocabSize: int, nbits: int):
  hashes = []
  for i in range(nbits):
    hashes.append(createHashFunction(vocabSize))
  return hashes

# creating 25 minhash vectors
minhashVectors = createMinhashFunction(len(vocab), 20)
print(minhashVectors)
```

```python
# create signature

# Logic:
'''
for i in range(1, len(vocab) + 1):
  idA = hashEx.index(i)
  signatureVal = a1Hot[idA]
  print(f"{i} -> {idA} -> {signatureVal}")
  if signatureVal == 1:
    print('match!')
    break
'''
# funtion to create the signature using one hot coded lists
def produceHash(hotlist: list):
  signature = []
  for func in minhashVectors:
    for i in range(1, len(vocab)):
      idX = func.index(i)
      signatureValue = hotlist[idX]
      if signatureValue == 1:
        signature.append(idX)
  return signature

# create signature for different text/sentence inputs
signatureA = produceHash(a1Hot)
signatureB = produceHash(b1Hot)
signatureC = produceHash(c1Hot)
signatureD = produceHash(d1Hot)
print(signatureA)
```

```python
# similarity function
def jaccard(x, y):
  return len(x.intersection(y)) / len(x.union(y))

print(jaccard(set(signatureA), set(signatureB)))
print(jaccard(set(signatureB), set(signatureD)))
```
